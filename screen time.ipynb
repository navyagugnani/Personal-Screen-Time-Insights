```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Screentime-App-Details-Dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
most_unlocks = df['Times opened'].max()
print(most_unlocks)

corelated_notifications = df.loc[df['Times opened'] == most_unlocks, 'Notifications'].iloc[0]
print('No. of times user got notification = ', corelated_notifications, ' Max number of unlocks to their phone =', most_unlocks)
```

```python
most_notifications = df['Notifications'].max()
corelated_unlocks = df.loc[df['Notifications'] == most_notifications, 'Times opened'].iloc[0]
print('No. of times user unlocked phone = ', corelated_unlocks, ' No. of max notifications =', most_notifications)
```

```python
df['Date']= df['Date'].astype("datetime64[ns]")
```

```python
plt.style.use("dark_background") 
df.plot(subplots=True, figsize=(12, 7))
plt.xlabel('Values')
```

```python
sorted_df = df.sort_values(by=['Usage'], ascending = False)
sorted_df.head()
```

```python
plt.figure(figsize=(15,8))
plt.style.use("dark_background")
pink_palette = sns.light_palette("green", n_colors=20)

sns.barplot(x='Date', y='Usage', data=sorted_df.head(20), palette=pink_palette)

plt.yticks(fontsize=6)
plt.ylabel('Usage')
plt.xlabel('Date')
plt.xticks(rotation=60)
plt.title('Total Usage per Day')
plt.tight_layout()
plt.show()
```

```python
apps = df['App'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(apps.index, apps, color=['green', 'lightgreen']) 
plt.xticks(rotation = 60)
plt.show()
```
